```python
"""
Given a fixed-length integer array arr, duplicate each occurrence of zero, shifting the remaining elements to the right.

Note that elements beyond the length of the original array are not written. 
Do the above modifications to the input array in place and do not return anything.

Example:

Input: arr = [1,0,2,3,0,4,5,0]
Output: [1,0,0,2,3,0,0,4]
Explanation: After calling your function, the input array is modified to: [1,0,0,2,3,0,0,4]

"""
```

```python
class Solution():
    def duplicates(self, arr):
        
        #Find count of zeros
        zeros = arr.count(0)
        
        #Edge cases
        if zeros == len(arr) or zeros == 0:
            pass
        
        #Declare length var, index pointer for list
        else:
            length = len(arr)
            i = 0
        
        #Test print
        #print(zeros)
        
        #While length less than index pointer
        while (length > i):
            
            #If index is zero, insert to the right, pop last element, skip over the duplicate zero with index + 2
            if (arr[i] == 0):
                arr.insert(i + 1, 0)
                arr.pop()
                i += 2
            #If index is not zero, increment index + 1, keep searching
            else:
                i += 1
                
        print(arr)
            
                
                
        
arr = arr = [1,0,2,3,0,4,5,0]
        
sol = Solution()
sol.duplicates(arr)
```
